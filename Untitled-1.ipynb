```python
import numpy as np
import scipy.sparse as sparse
from scipy.sparse.linalg import eigsh
import itertools
from collections import defaultdict

# --- System Parameters ---
L = 4  # Lattice dimension (4x4)
N_sites = L * L
N_particles = 4
BC_TYPE = "Periodic"

# --- Model Parameters ---
t = 1.0  # Hopping strength
V = 200000.0  # Nearest-neighbor interaction strength
         # V > 0: Repulsive
         # V < 0: Attractive
         # V = 0: Non-interacting

def get_neighbors(site, L):
    """
    Gets nearest neighbors for a site on an L-by-L grid
    with Periodic Boundary Conditions (PBC).
    """
    neighbors = []
    row, col = divmod(site, L)
    
    # Up (wraps around)
    up_row = (row - 1) % L
    neighbors.append(up_row * L + col)
    
    # Down (wraps around)
    down_row = (row + 1) % L
    neighbors.append(down_row * L + col)
    
    # Left (wraps around)
    left_col = (col - 1) % L
    neighbors.append(row * L + left_col)
    
    # Right (wraps around)
    right_col = (col + 1) % L
    neighbors.append(row * L + right_col)
        
    return neighbors

def are_neighbors(site1, site2, L):
    """
    Checks if two sites are nearest neighbors.
    This function relies on get_neighbors and works for
    both OBC and PBC without modification.
    """
    return site2 in get_neighbors(site1, L)

def calculate_fermionic_sign(state, k_idx, j):
    """
    Calculates the new state and the fermionic sign for a hop.
    
    Args:
        state (tuple): The current sorted tuple of occupied sites.
        k_idx (int): The index *within the tuple* of the particle that is hopping.
        j (int): The destination site index (0 to N_sites-1).
        
    Returns:
        (tuple, int): (new_state, sign)
    """
    
    # Create the new state by removing k and adding j
    state_list = list(state)
    state_list.pop(k_idx)
    state_list.append(j)
    
    # The sign is determined by the number of swaps to re-sort the list.
    # 1. Sign from removing k: (-1)^(k_idx)
    sign = (-1)**k_idx
    
    # 2. Find where j lands in the new sorted list
    # We can just sort and find the new index of j
    state_list.sort()
    j_idx = state_list.index(j)
    
    # 3. Sign from inserting j: (-1)^(j_idx)
    sign *= (-1)**j_idx
    
    new_state = tuple(state_list)
    return new_state, sign

def print_lattice_state(state, L):
    """Prints a visual representation of the lattice state."""
    grid = [['.' for _ in range(L)] for _ in range(L)]
    for site in state:
        row, col = divmod(site, L)
        grid[row][col] = 'F'
    
    print("  " + " ".join([f"{i}" for i in range(L)]))
    print(" +" + "-" * (2*L - 1))
    for r, row in enumerate(grid):
        print(f"{r}|" + " ".join(row))
    print("")

def build_and_diagonalize():
    """
    Builds the basis, constructs the Hamiltonian, and finds the ground state.
    *** MODIFIED TO HANDLE DEGENERACY ***
    """
    print(f"System: {L}x{L} lattice, {N_particles} spinless fermions")
    print(f"Boundary Conditions: {BC_TYPE}")
    print(f"Parameters: t={t}, V={V}\n")

    # === 1. Build the Basis ===
    print("Building basis...")
    sites = range(N_sites)
    # The basis is all combinations of N_particles on N_sites
    basis = list(itertools.combinations(sites, N_particles))
    dim = len(basis)
    
    # Create a reverse-lookup map for state -> index
    state_to_index = {state: i for i, state in enumerate(basis)}
    
    print(f"Hilbert space dimension: {dim} (binom({N_sites}, {N_particles}))\n")

    # === 2. Build the Hamiltonian Matrix ===
    print("Building Hamiltonian matrix...")
    H = sparse.lil_matrix((dim, dim), dtype=np.float64)
    
    for i, state in enumerate(basis):
        # --- Diagonal Part (H_V) ---
        E_V = 0.0
        for (site1, site2) in itertools.combinations(state, 2):
            if are_neighbors(site1, site2, L):
                E_V += V
        H[i, i] = E_V
        
        # --- Off-Diagonal Part (H_t) ---
        for k_idx, k in enumerate(state):  # k is the occupied site
            for j in get_neighbors(k, L): # j is the adjacent neighbor
                if j not in state: # Hop only to EMPTY sites
                    new_state, sign = calculate_fermionic_sign(state, k_idx, j)
                    f = state_to_index[new_state]
                    H[f, i] = -t * sign

    print("Hamiltonian build complete.")
    H_csr = H.tocsr()

    # === 3. Diagonalize ===
    print("Diagonalizing to find ground state(s) (using sparse 'eigsh')...")
    
    # *** MODIFIED: Ask for more eigenvalues to check for degeneracy ***
    # We ask for k=10 lowest eigenvalues (or k=dim if dim is very small)
    num_eigenvalues_to_find = min(10, dim - 1)
    
    try:
        eigenvalues, eigenvectors = eigsh(H_csr, k=num_eigenvalues_to_find, which='SA')
    except Exception as e:
        print(f"Error during diagonalization: {e}")
        # Fallback for small matrices
        eigenvalues, eigenvectors = np.linalg.eigh(H_csr.toarray())


    # --- NEW: Check for degeneracy ---
    gs_energy = eigenvalues[0]
    tolerance = 1e-4 # Numerical tolerance for checking equality
    
    # Find all indices where the eigenvalue is close to the ground state energy
    degenerate_indices = np.where(np.abs(eigenvalues - gs_energy) < tolerance)[0]
    num_degenerate = len(degenerate_indices)

    print("\n" + "="*40)
    print(f"         Ground State Energy: {gs_energy:.8f}")
    if num_degenerate > 1:
        print(f"   !!! Ground State is {num_degenerate}-fold DEGENERATE !!!")
        print(f"   All {num_degenerate} states in this energy level will be shown.")
    print("="*40)


    # === 4. Analyze the Ground State Vector(s) ===
    # --- MODIFIED: Loop over all degenerate states ---
    
    for n in range(num_degenerate):
        
        # Get the n-th degenerate ground state vector
        state_index = degenerate_indices[n]
        gs_vector = eigenvectors[:, state_index]
        gs_e = eigenvalues[state_index] # The specific energy for this vector

        print(f"\n--- Analyzing Degenerate Ground State {n + 1} / {num_degenerate} (E = {gs_e:.8f}) ---")
        print("Showing top 10 most probable basis states.\n")
        
        # Calculate probabilities |C_i|^2
        probabilities = np.abs(gs_vector)**2
        
        # Get the indices that would sort the probabilities in descending order
        sorted_indices = np.argsort(probabilities)[::-1]
        
        num_components_shown = 0
        for i in range(max(100, dim)):
            idx = sorted_indices[i]
            prob = probabilities[idx]
            
            # Don't show components with ~zero probability
            if prob < tolerance:
                continue 
                
            coeff = gs_vector[idx]
            state = basis[idx]
            
            print(f"--- Component {num_components_shown + 1} ---")
            print(f"  Probability: {prob*100:6.2f}%")
            print(f"  Coefficient: {coeff: 9.6f}")
            print(f"  State (occupied sites): {state}")
            print_lattice_state(state, L)
            num_components_shown += 1
            
        if num_components_shown == 0:
            print("This state vector has no components above the numerical tolerance.")
            
        print("-" * 20) # Separator for clarity


# --- Run the calculation ---
if __name__ == "__main__":
    build_and_diagonalize()
```

```python
import numpy as np
import scipy.sparse as sparse
from scipy.sparse.linalg import eigsh
import itertools
from collections import defaultdict
import matplotlib.pyplot as plt

# --- System Parameters ---
L = 4  # Lattice dimension (4x4)
N_sites = L * L
N_particles = 4
BC_TYPE = "Periodic"

# --- Model Parameters ---
t = 1.0  # Hopping strength
V = 999999.0  # Nearest-neighbor interaction strength
         # V > 0: Repulsive
         # V < 0: Attractive
         # V = 0: Non-interacting

def get_neighbors(site, L):
    neighbors = []
    row, col = divmod(site, L)
    up_row = (row - 1) % L
    neighbors.append(up_row * L + col)
    down_row = (row + 1) % L
    neighbors.append(down_row * L + col)
    left_col = (col - 1) % L
    neighbors.append(row * L + left_col)
    right_col = (col + 1) % L
    neighbors.append(row * L + right_col)
    return neighbors

def are_neighbors(site1, site2, L):
    return site2 in get_neighbors(site1, L)

def calculate_fermionic_sign(state, k_idx, j):
    state_list = list(state)
    state_list.pop(k_idx)
    state_list.append(j)
    sign = (-1)**k_idx
    state_list.sort()
    j_idx = state_list.index(j)
    sign *= (-1)**j_idx
    new_state = tuple(state_list)
    return new_state, sign

def print_lattice_state(state, L):
    grid = [['.' for _ in range(L)] for _ in range(L)]
    for site in state:
        row, col = divmod(site, L)
        grid[row][col] = 'F'
    print("  " + " ".join([f"{i}" for i in range(L)]))
    print(" +" + "-" * (2*L - 1))
    for r, row in enumerate(grid):
        print(f"{r}|" + " ".join(row))
    print("")

def visualize_ground_state_correlations(gs_vector, basis, L, state_label="GS"):
    """
    Calculates and plots the DENSITY correlations.
    (This is from the previous step, but still very useful!)
    """
    print(f"\n--- Visualizing DENSITY Correlations for {state_label} ---")
    N_sites = L * L
    density = np.zeros(N_sites)
    correlator_vs_site_0 = np.zeros(N_sites)
    
    for k, state in enumerate(basis):
        prob = np.abs(gs_vector[k])**2
        if prob < 1e-12: continue
        for i in state:
            density[i] += prob
        if 0 in state:
            for j in state:
                correlator_vs_site_0[j] += prob
                
    if density[0] > 1e-9:
        conditional_prob = correlator_vs_site_0 / density[0]
    else:
        conditional_prob = correlator_vs_site_0
        
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))
    fig.suptitle(f"Density Visualization ({state_label}) | V={V}, t={t}", fontsize=16)
    
    im1 = axes[0].imshow(density.reshape((L, L)), cmap='viridis', interpolation='nearest')
    axes[0].set_title(r"Average Density $\langle n_i \rangle$")
    fig.colorbar(im1, ax=axes[0])

    im2 = axes[1].imshow(conditional_prob.reshape((L, L)), cmap='inferno', interpolation='nearest')
    axes[1].set_title(r"Conditional Probability $P(j | 0) = \langle n_0 n_j \rangle / \langle n_0 \rangle$")
    axes[1].text(0, 0, '0', color='cyan', ha='center', va='center', fontsize=16, weight='bold')
    fig.colorbar(im2, ax=axes[1])

    for ax in axes:
        ax.set_xticks(np.arange(L)); ax.set_yticks(np.arange(L))
        ax.set_xticklabels(np.arange(L)); ax.set_yticklabels(np.arange(L))

    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()

# ******************************************************
# *** NEW FUNCTIONS TO VISUALIZE BOND ORDER ***
# ******************************************************

def calculate_bond_expectation(gs_vec, basis, state_to_index, bond):
    """
    Calculates <gs| (c_i^dag c_j + c_j^dag c_i) |gs> for a specific bond.
    """
    i, j = bond
    K_ij = 0.0
    
    for k_idx, k_state in enumerate(basis):
        C_k = gs_vec[k_idx]
        if np.abs(C_k) < 1e-9:
            continue
            
        # Term 1: <gs| c_i^dag c_j |gs>
        # We need state k to have j but not i
        if j in k_state and i not in k_state:
            j_idx_in_tuple = k_state.index(j)
            l_state, sign = calculate_fermionic_sign(k_state, j_idx_in_tuple, i)
            l_idx = state_to_index[l_state]
            C_l = gs_vec[l_idx]
            K_ij += np.conj(C_l) * C_k * sign
            
        # Term 2: <gs| c_j^dag c_i |gs>
        # We need state k to have i but not j
        if i in k_state and j not in k_state:
            i_idx_in_tuple = k_state.index(i)
            l_state, sign = calculate_fermionic_sign(k_state, i_idx_in_tuple, j)
            l_idx = state_to_index[l_state]
            C_l = gs_vec[l_idx]
            K_ij += np.conj(C_l) * C_k * sign
            
    # Since gs_vec is real for this Hamiltonian, K_ij is real
    return np.real(K_ij)

def visualize_bond_correlations(gs_vector, basis, state_to_index, L, state_label="GS"):
    """
    Calculates and plots the bond-order correlator <c_i^dag c_j + h.c.>
    for all horizontal and vertical bonds.
    """
    print(f"\n--- Visualizing BOND Correlations (Dimer Basis) for {state_label} ---")
    
    horizontal_bonds = np.zeros((L, L))
    vertical_bonds = np.zeros((L, L))
    
    for r in range(L):
        for c in range(L):
            j_site = r * L + c
            
            # Horizontal bond: (r, c) --- (r, c+1)
            i_site_h = r * L + (c + 1) % L
            bond_h = tuple(sorted((j_site, i_site_h)))
            # We store the bond at site (r, c)
            horizontal_bonds[r, c] = calculate_bond_expectation(
                gs_vector, basis, state_to_index, bond_h
            )
            
            # Vertical bond: (r, c) --- (r+1, c)
            i_site_v = ((r + 1) % L) * L + c
            bond_v = tuple(sorted((j_site, i_site_v)))
            # We store the bond at site (r, c)
            vertical_bonds[r, c] = calculate_bond_expectation(
                gs_vector, basis, state_to_index, bond_v
            )

    fig, axes = plt.subplots(1, 2, figsize=(13, 5))
    fig.suptitle(f"Bond-Order (Dimer) Visualization ({state_label}) | V={V}, t={t}", fontsize=16)

    # Plot 1: Horizontal Bonds
    # A value at (r,c) is the strength of the bond (r,c)---(r,c+1)
    im1 = axes[0].imshow(horizontal_bonds, cmap='Spectral', interpolation='nearest')
    axes[0].set_title(r"Horizontal Bonds $\langle K_{i, i+x} \rangle$")
    fig.colorbar(im1, ax=axes[0])

    # Plot 2: Vertical Bonds
    # A value at (r,c) is the strength of the bond (r,c)---(r+1,c)
    im2 = axes[1].imshow(vertical_bonds, cmap='Spectral', interpolation='nearest')
    axes[1].set_title(r"Vertical Bonds $\langle K_{i, i+y} \rangle$")
    fig.colorbar(im2, ax=axes[1])

    for ax in axes:
        ax.set_xticks(np.arange(L)); ax.set_yticks(np.arange(L))
        ax.set_xticklabels(np.arange(L)); ax.set_yticklabels(np.arange(L))

    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()

# ******************************************************
# *** Main calculation function (modified to call BOTH plots) ***
# ******************************************************
def build_and_diagonalize():
    """
    Builds the basis, constructs the Hamiltonian, and finds the ground state.
    """
    print(f"System: {L}x{L} lattice, {N_particles} spinless fermions")
    print(f"Boundary Conditions: {BC_TYPE}")
    print(f"Parameters: t={t}, V={V}\n")

    # === 1. Build the Basis ===
    print("Building basis...")
    sites = range(N_sites)
    basis = list(itertools.combinations(sites, N_particles))
    dim = len(basis)
    state_to_index = {state: i for i, state in enumerate(basis)}
    print(f"Hilbert space dimension: {dim}\n")

    # === 2. Build the Hamiltonian Matrix ===
    print("Building Hamiltonian matrix...")
    H = sparse.lil_matrix((dim, dim), dtype=np.float64)
    for i, state in enumerate(basis):
        E_V = 0.0
        for (site1, site2) in itertools.combinations(state, 2):
            if are_neighbors(site1, site2, L):
                E_V += V
        H[i, i] = E_V
        for k_idx, k in enumerate(state):
            for j in get_neighbors(k, L):
                if j not in state:
                    new_state, sign = calculate_fermionic_sign(state, k_idx, j)
                    f = state_to_index[new_state]
                    H[f, i] = -t * sign
    print("Hamiltonian build complete.")
    H_csr = H.tocsr()

    # === 3. Diagonalize ===
    print("Diagonalizing to find ground state(s)...")
    num_eigenvalues_to_find = min(10, dim - 1)
    try:
        eigenvalues, eigenvectors = eigsh(H_csr, k=num_eigenvalues_to_find, which='SA')
    except:
        eigenvalues, eigenvectors = np.linalg.eigh(H_csr.toarray())

    gs_energy = eigenvalues[0]
    tolerance = 1e-4
    degenerate_indices = np.where(np.abs(eigenvalues - gs_energy) < tolerance)[0]
    num_degenerate = len(degenerate_indices)

    print("\n" + "="*40)
    print(f"         Ground State Energy: {gs_energy:.8f}")
    if num_degenerate > 1:
        print(f"   !!! Ground State is {num_degenerate}-fold DEGENERATE !!!")
    print("="*40)

    # === 4. Analyze and Visualize the Ground State(s) ===
    for n in range(num_degenerate):
        state_index = degenerate_indices[n]
        gs_vector = eigenvectors[:, state_index]
        gs_e = eigenvalues[state_index]
        label = f"GS {n + 1}/{num_degenerate}"
        print(f"\n--- Analyzing Degenerate Ground State {n + 1} / {num_degenerate} (E = {gs_e:.8f}) ---")
        
        # --- Print top components ---
        probabilities = np.abs(gs_vector)**2
        sorted_indices = np.argsort(probabilities)[::-1]
        print("Showing top 5 most probable basis states:")
        for i in range(min(5, dim)):
            idx = sorted_indices[i]
            prob = probabilities[idx]
            if prob < tolerance: continue
            print(f"  {i+1}. Prob: {prob*100:6.2f}%, Coeff: {gs_vector[idx]: 9.6f}, State: {basis[idx]}")
        
        # --- Call Density visualization ---
        visualize_ground_state_correlations(gs_vector, basis, L, state_label=label)
        
        # --- Call NEW Bond visualization ---
        visualize_bond_correlations(gs_vector, basis, state_to_index, L, state_label=label)

# --- Run the calculation ---
if __name__ == "__main__":
    build_and_diagonalize()
```
